```python
# importing the RoBERTa model and the tokenizer
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

tokenizer = AutoTokenizer.from_pretrained('cardiffnlp/twitter-roberta-base-sentiment-latest')
model = AutoModelForSequenceClassification.from_pretrained('cardiffnlp/twitter-roberta-base-sentiment-latest',from_tf=True)
```

```python
# function to tokenize and give output as Sentiment
def sentiment_score(review):
  tokens = tokenizer.encode(review, return_tensors='pt')
  result = model(tokens)
  return int(torch.argmax(result.logits))
```

```python
# applying the model and making a new column as sentiment
dfSentiment['sentiment'] = dfSentiment['new_tweet'].dropna().apply(sentiment_score)
```
